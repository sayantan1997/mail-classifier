```python
#creating a dictionary holding count of each word in the train set
import collections
allwords=[]
for i in range(0,len(x_train)):
    words=x_train["content"][i].split()
    allwords+=words
dictionary1=collections.Counter(allwords)
k=list(dictionary1.keys()) #convert to list else would result  in Runtime error 
for i in k:
    if i.isalpha()==False:
        del dictionary1[i]
    if len(i)==1:
        del dictionary1[i]
```

```python
#most common 3000 words
dictionary1=dictionary1.most_common(3000)
#for j in dictionary1.keys():
#    print(j)
```

```python
#creating frequency matrices
def create(file):
    features_matrix4=np.zeros((len(file),3000))
    for i in range(0,len(file)):
        words=file["content"][i].split()
        for p,j in enumerate(dictionary1):
            for word in words:
                if word==j[0]:
                    features_matrix4[i,p]=words.count(word)
                    
    return(features_matrix4)
```
